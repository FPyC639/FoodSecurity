# Food Insecurity Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("h240.xlsx")
```

```python
print(f"The columns for this data set are:{chr(10)}{f'{chr(9)}{chr(10)}'.join(df.columns)}")
```

```python
def func1(i):
  match i:
    case 2:
      return "No"
    case -1:
      return "Not Applicable"
    case -8:
      return "Don't Know"
    case 1:
      return "Yes"
```

```python
def func2(i):
  match i:
    case -8:
      return "Don't Know"
    case 1:
      return "Often True"
    case 2:
      return "Sometimes True"
    case 3:
      return "Never True"
```

```python
plt.figure(figsize=(10,10))
ax = df.FSHGRY42.apply(lambda x: func1(x)) \
      .value_counts() \
      .plot.bar(rot=45,
                title= "Did you ever go hungry?",
                xlabel= "Responses",
                ylabel = "Values")
plt.savefig("Didyouevergohungry.png")
```

```python
plt.figure(figsize=(10,10))
ax1 = df.FSLAST42.apply(lambda x: func2(x)) \
      .value_counts() \
      .plot.bar(rot=45,
                title = "How often did food not last?",
                xlabel= "Reponses",
                ylabel= "Values")
plt.savefig("Howoftendidfoodnotlast.png")
```
